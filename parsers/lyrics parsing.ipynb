```python
def url_artist(artist):
    #link for all artists with N-letter
    first_letter = artist[0].lower()
    url = 'https://www.azlyrics.com/' + first_letter + '.html'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    
    #link for finding artist
    for link in soup.find_all('div', class_='col-sm-6 text-center artist-col'):
        for artist_name in link.findAll("a", href=True):
            if artist.lower() == artist_name.text.lower():
                url = 'https://www.azlyrics.com/' + artist_name.get('href')
                         
    return url
```

```python
def find_song(artist, song):
    #link for all artists with N-letter
    first_letter = artist[0].lower()
    url = 'https://www.azlyrics.com/' + first_letter + '.html'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    
    #link for finding artist
    for link in soup.find_all('div', class_='col-sm-6 text-center artist-col'):
        for artist_name in link.findAll("a", href=True):
            if artist.lower() == artist_name.text.lower():
                url = 'https://www.azlyrics.com/' + artist_name.get('href')
    
    
    
    #link for finding song
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    for link in soup.find_all('div', class_='listalbum-item'):
        for song_name in link.findAll("a", href=True, target="_blank"):
            if song.lower() == song_name.text.lower():
                url = 'https://www.azlyrics.com/' + song_name.get('href')
             
                
                
    #getting lyrics           
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    lyrics = soup.select('div')[22].get_text()
    lyrics = re.sub('(\\r|)(\\n)', ' ', str(lyrics))
    if lyrics == ' Hozier - "Swan Upon Leda" Ed Sheeran - "Celestial" Lil Baby - "Heyy" SEULGI - "28 Reasons" GloRilla & Cardi B - "Tomorrow 2" ':
        lyrics = ''
    
    return lyrics
    
```

```python
class Music:
    
    def __init__(self, artist):
        self.artist = artist
        self.url_artist = ''
        self.__url_song = ''
        self.lyrics = ''
        self.song = ''
        
    def get_songs(self):
        #link for all artists with N-letter
        first_letter = self.artist[0].lower()
        url = 'https://www.azlyrics.com/' + first_letter + '.html'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'lxml')
    
        #link for finding artist
        for link in soup.find_all('div', class_='col-sm-6 text-center artist-col'):
            for artist_name in link.findAll("a", href=True):
                if self.artist.lower() == artist_name.text.lower():
                    self.url_artist = 'https://www.azlyrics.com/' + artist_name.get('href')
        print(self.url_artist)

    def get_lyrics(self, song):
        response = requests.get(self.url_artist)
        soup = BeautifulSoup(response.text, 'lxml')
        #link for finding song
        for link in soup.find_all('div', class_='listalbum-item'):
            for song_name in link.findAll("a", href=True, target="_blank"):
                if song.lower() == song_name.text.lower():
                    self.__url_song = 'https://www.azlyrics.com/' + song_name.get('href')
        
        #getting lyrics           
        response = requests.get(self.__url_song)
        soup = BeautifulSoup(response.text, 'lxml')
        self.lyrics = soup.select('div')[22].get_text()
        self.lyrics = re.sub('(\\r|)(\\n)', ' ', str(self.lyrics))
        if self.lyrics == ' Hozier - "Swan Upon Leda" Ed Sheeran - "Celestial" Lil Baby - "Heyy" SEULGI - "28 Reasons" GloRilla & Cardi B - "Tomorrow 2" ':
            self.lyrics = 'Lyrics not found'
        else:
            print(self.lyrics)
```

```python
def play_song(artist):
    #link for all artists with N-letter
    first_letter = artist[0].lower()
    url = 'https://musify.club/artist/letter-' + first_letter 
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    
   
    pages = soup.find('ul', class_='pagination justify-content-center')
    urls = []
    links = pages.find_all('a', class_='page-link')

    #получем ссылки каждой страницы
    for link in links:
        pageNum = int(link.text) if link.text.isdigit() else None
        if pageNum != None:
            hrefval = link.get('href')
            urls.append(hrefval)
        
       
     #меняем страницы 
    for i in range(2000):
        newUrl = url.replace('page2', f'page{i}')
        response = requests.get(newUrl)
        soup = BeautifulSoup(response.text, 'lxml')
    
        #link for finding artist
        for link in soup.find_all('div', class_='col-md-8'):
            for artist_name in link.findAll("a", href=True):
                if artist.lower() == artist_name.text.lower():
                    url = 'https://musify.club/' + artist_name.get('href')

    
    return url
```
